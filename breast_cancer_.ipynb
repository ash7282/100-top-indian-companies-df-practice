```python
# prompt: compaire diagnosis columns unique value with other data on perfect comparision stats

for col in df.columns:
  if col != 'diagnosis':
    unique_values = df['diagnosis'].unique()
    for value in unique_values:
      df_subset = df[df['diagnosis'] == value]
      print(f"Comparison for diagnosis value {value}:")
      print(f"Mean of {col}: {df_subset[col].mean()}")
      print(f"Standard deviation of {col}: {df_subset[col].std()}")
      print(f"Minimum value of {col}: {df_subset[col].min()}")
      print(f"Maximum value of {col}: {df_subset[col].max()}")
      print()
```

```python
# prompt: compaire diagnosis columns unique value with other data on perfect comparision visualization

import matplotlib.pyplot as plt
import seaborn as sns

for col in df.columns:
  if col != 'diagnosis':
    unique_values = df['diagnosis'].unique()
    for value in unique_values:
      df_subset = df[df['diagnosis'] == value]
      plt.figure(figsize=(10, 6))
      sns.histplot(df_subset[col], bins=10, label=f"Diagnosis: {value}")
      plt.xlabel(col)
      plt.ylabel("Frequency")
      plt.legend()
      plt.title(f"Distribution of {col} for Diagnosis { value} ")
      plt.show()
```

```python
# prompt: datapoints plot with respect to diagnosis columns

import matplotlib.pyplot as plt
# Create a figure and axes object
fig, ax = plt.subplots()

# Iterate through each column in the DataFrame
for column in df.columns:
  # If the column is not the diagnosis column
  if column != 'diagnosis':
    # Plot the datapoints for each diagnosis value
    for diagnosis_value in df['diagnosis'].unique():
      df_subset = df[df['diagnosis'] == diagnosis_value]
      ax.scatter(df_subset[column], df_subset['diagnosis'], label=diagnosis_value)

# Set the title and axis labels
ax.set_title('Datapoints Plot with Respect to Diagnosis')
ax.set_xlabel('Column Values')
ax.set_ylabel('Diagnosis')

# Add a legend and show the plot
ax.legend()
plt.show()
```

```python
df.columns=df.columns.str.strip()
```

```python
X=df.drop(['diagnosis'],axis=1)
y=df['diagnosis']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
print(f'X_train {X_train.shape}')
print(f'X_test {X_test.shape}')
print(f'y_train {y_train.shape}')
print(f'y_test  {y_test .shape}')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
logreg.fit(X_train,y_train)
```

```python
logreg_y_pred_train= logreg.predict(X_train)
logreg_y_pred_test = logreg.predict(X_test)
```

```python
from sklearn.metrics import f1_score
```

```python
print('Training f1 score ',f1_score(y_train,logreg_y_pred_train))
print('Testining f1 score ',f1_score(y_test,logreg_y_pred_test))
```
